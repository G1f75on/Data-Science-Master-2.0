```python
"""Q1.
GET: The GET method is used to request data from the server. It appends the data to the URL in the form of query parameters. 
This method is used for retrieving data and should not be used for actions that modify server data because it's not secure. 
For example, when you open a web page or click a link, your browser sends a GET request to the server to fetch the page's content.

POST: The POST method is used to submit data to the server to create or update resources. Instead of appending data to the URL, 
it sends the data in the request body. This method is more secure for sensitive data like passwords, as the data is not visible in the URL.
It's commonly used in forms on websites to send user input to the server for processing.


Q2.
In Flask, request is a global object provided by the Flask framework that encapsulates the information about the current HTTP request 
made by a client (e.g., a web browser) to the Flask application. It allows you to access various parts of the request, such as the form 
data, URL parameters, headers, cookies, and the request method (e.g., GET, POST).

By using request, you can extract data from the client's request and use it to make decisions or perform actions within your Flask 
application. For example, you can access form data submitted by a user, retrieve query parameters, or check if the request is a POST 
request to handle form submissions.


Q3.
In Flask, redirect() is a function used to perform URL redirection. It allows you to redirect the client's browser from one URL to another.
Redirects are useful when you want to redirect users to a different page after they perform a certain action on your website or when you 
want to ensure that the user visits a specific URL.


Q4. 
In Flask, templates are files containing dynamic content written in a markup language such as HTML, with the ability to embed dynamic
data into the template. They allow you to separate the presentation layer from the business logic in your Flask application.

The render_template() function in Flask is used to render these templates and inject dynamic data into them. It takes the name of the 
template file as a parameter and returns the template's content with the provided data included. This function is often used within Flask 
route functions to generate HTML pages dynamically based on the data retrieved from the server or user inputs.
"""
```
