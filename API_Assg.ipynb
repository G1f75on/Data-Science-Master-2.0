```python
""" Q1. 
API stands for Application Programming Interface. It is a set of rules and protocols that allows different software applications to communicate 
and interact with each other. APIs define the methods and data formats that applications can use to request and exchange information. They enable 
developers to access the functionality of a system or service without having to understand its internal workings.

Example of API usage in real life: UPI Banking

When you sign in to an UPI account like GPay, PayTM, you are using an API. In this scenario, the app 
relies on the Bank API to authenticate your identity and grant access. The API enables the seamless transactions and of user information 
between the app and Bank, simplifying the transactional process for users.


Q2.
Advantages of using API:

Interoperability: APIs enable different software systems to work together, promoting interoperability and integration.
Reusability: APIs allow developers to reuse existing functionalities, saving time and effort in development.
Scalability: APIs help in building scalable applications by distributing tasks across multiple services.
Rapid Development: APIs speed up development since developers can focus on building features rather than creating everything from scratch.
Improved User Experience: APIs facilitate the integration of third-party services, enhancing the user experience with more features and options.
Disadvantages of using API:

Dependency: When an API is used, the application becomes dependent on the API provider, and any changes or downtime in the API can affect the application's functionality.
Security Risks: Poorly implemented APIs can be susceptible to security vulnerabilities, leading to potential breaches if not properly secured.
Complexity: APIs can add complexity to the application, especially when dealing with multiple APIs and integrating different services.
Performance: Overuse or inefficient use of APIs can negatively impact the application's performance and response times.


Q3.
A Web API (Web Application Programming Interface) is an API specifically designed to be accessed over the internet through HTTP requests. 
It allows communication between different web applications and services. Web APIs follow specific architectural principles, such as REST and SOAP, 
to define the rules and conventions for interaction.

Difference between API and Web API:

API (Application Programming Interface) is a broader term that encompasses all types of interfaces used for communication between software components,
whether they are web-based or not.
Web API specifically refers to APIs accessible over the internet using HTTP/HTTPS protocols. It is a type of API designed for web-based communication.



Q4. 
REST Architecture:

REST is an architectural style that uses standard HTTP methods (GET, POST, PUT, DELETE) to perform CRUD (Create, Read, Update, Delete) operations on 
resources.
It operates over the existing HTTP protocol, making it simple and lightweight.
RESTful APIs use URLs (Uniform Resource Locators) to represent resources, and responses are often sent in JSON or XML format.
REST is stateless, meaning each request from a client to the server must contain all the information needed to understand and process the request.
SOAP Architecture:

SOAP is a protocol and an architectural style that defines the XML-based message format for exchanging structured information in the implementation 
of web services.
It relies on XML for message format, SOAP envelopes, and headers.
SOAP uses HTTP, SMTP, TCP, and other protocols for message transport, allowing it to work across different network protocols.
It provides more extensive standards for security, error handling, and message routing.
Shortcomings of SOAP:

Complexity: SOAP messages are relatively complex due to XML formatting, making them bulkier and more challenging to parse.
Performance: SOAP messages require more bandwidth and processing time compared to RESTful APIs due to their larger size and parsing complexity.
Lack of native support: Not all platforms or programming languages have built-in support for SOAP, making its implementation more challenging in some 
cases. Limited human readability: SOAP messages in XML format can be challenging for humans to read and understand compared to JSON used by REST APIs.

""
```
