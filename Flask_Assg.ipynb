```python
"""Q5.
The url_for() function in Flask is used for URL building. It is a helper function that generates a URL for the specified view function, 
taking into account any dynamic parts of the URL and the routing rules defined in the application.
"""
```

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    details_url = url_for('details')
    
    welcome_url = url_for('welcome')
    
    return f"Welcome to our website! Check out the details <a href='{details_url}'>here</a> or the welcome message <a href='{welcome_url}'>here</a>."

@app.route('/welcome')
def welcome():
    return "Welcome to our website! We are glad you are here."

@app.route('/details')
def details():
    return "Company Name: ABC Corporation<br>Location: India<br>Contact Detail: 999-999-9999"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
