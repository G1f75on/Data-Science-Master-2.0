```python
"Scatter plot, is a chart that shows the relationship between two variables."
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Scatter Plot")
```

```python
"Subplots mean a group of smaller axes that can exist together within a single figure."
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
y1 = np.array([50, 20, 40, 20, 60, 70])
y2 = np.array([10, 20, 30, 40, 50, 60])
y3 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot(x,y)
ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
```

```python
"Bar plots are a type of data visualization used to represent data in the form of rectangular bars. "
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit, color = 'pink')
plt.xlabel("Companies")
plt.ylabel("Profit")
plt.title("Company's Profit")
```

```python
plt.barh(company,profit, color = 'r')
plt.xlabel("Profit")
plt.ylabel("Companies")
plt.title("Company's Profit")
```
